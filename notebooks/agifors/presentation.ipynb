```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

# Exponential demand curve

```python
ps = list(range(1500, 2500, 100))
```

```python
V1, w1 = 50, 500
```

```python
V2, w2 = 55, 500
```

```python
ds1 = [V1 * np.exp(-(p - 1500)/w1) for p in ps]
ds2 = [V2 * np.exp(-(p - 1500)/w2) for p in ps]
```

```python
demand1 = go.Scatter(
    x=ps,
    y=ds1,
    mode="lines",
    name="V=50",
    line=dict(color="#ffa500")
)

demand2 = go.Scatter(
    x=ps,
    y=ds2,
    mode="lines",
    name="V=55",
    line=dict(color="#ff0000")
)

fig = go.Figure(data=[demand1, demand2])
fig.update_layout(
    xaxis_rangeslider_visible=False,
    width=1000,
    height=500
)
fig.show()
```

# Piecewise-linear demand curve

```python
ps = [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200, 2300, 2400]
ds1 = [50, 42, 33, 27, 22, 18, 15, 12, 10, 8]
ds2 = [55, 46, 37, 30, 25, 21, 17, 14, 11, 9]
```

```python
demand1 = go.Scatter(
    x=ps,
    y=ds1,
    mode="lines+markers",
    name="V=50",
    line=dict(color="#ffa500")
)

demand2 = go.Scatter(
    x=ps,
    y=ds2,
    mode="lines+markers",
    name="V=55",
    line=dict(color="#ff0000")
)

fig = go.Figure(data=[demand1, demand2])
fig.update_layout(
    xaxis_rangeslider_visible=False,
    width=1000,
    height=500
)
fig.show()
```
