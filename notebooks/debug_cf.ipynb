```python
import pandas as pd 

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
fcstdate = "20240520"

depdates = ["20241001","20241002","20241003","20241004","20241005","20241006","20241007",
            "20241008","20241009","20241010","20241011","20241012","20241013","20241014",
            "20241015","20241016","20241017","20241018","20241019","20241020","20241021",
            "20241022","20241023","20241024","20241025","20241026","20241027","20241028",
            "20241029","20241030","20241031"]
```

```python
df = None
for depdate in depdates:
    fname = "s3://ay-emr-job/nrm/cf/{}/{}/{}/cf_{}_{}.csv.gz".format(fcstdate[:4], fcstdate[4:6], fcstdate[6:8], fcstdate, depdate)
    cf_df = pd.read_csv(fname, low_memory = False)
    if df is None:
        df = cf_df 
    else:
        df = pd.concat([df, cf_df])
```
