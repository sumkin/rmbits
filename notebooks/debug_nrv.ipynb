```python
import pandas as pd
from datetime import datetime as dt
from datetime import timedelta

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
fcstdate = "20240520"

depdates = []
depdate = "20240520"
while depdate <= "20250517":
    depdates.append(depdate)
    depdate = dt.strftime(dt.strptime(depdate, "%Y%m%d") + timedelta(days=1), "%Y%m%d")
```

# Read fdc files

```python
fdc_dfs = None
for depdate in depdates:
    fname = "s3://ay-emr-job/nrm/fdc/{}/{}/{}/fdc_{}_{}.csv.gz".format(fcstdate[:4], fcstdate[4:6], fcstdate[6:8], fcstdate, depdate)
    df = pd.read_csv(fname, low_memory = False)
    if fdc_dfs is None:
        fdc_dfs = [df] 
    else:
        fdc_dfs.append(df)
    print(depdate, end=" ")
fdc_df = pd.concat(fdc_dfs)
```

# Read constrained forecast

```python
cf_df = None
for depdate in depdates:
    fname = "s3://ay-emr-job/nrm/cf/{}/{}/{}/cf_{}_{}.csv.gz".format(fcstdate[:4], fcstdate[4:6], fcstdate[6:8], fcstdate, depdate)
    df = pd.read_csv(fname, low_memory = False)
    if cf_df is None:
        cf_df = df 
    else:
        cf_df = pd.concat([cf_df, df])
    print(depdate, end=" ")
```

```python
cf_df[(cf_df["BASE_OD_ORGN"] == "JFK") & (cf_df["BASE_OD_DSTN"] == "TLL") & (cf_df["POS"] == "US")].head(5)
```

# Read NRV PWL values

```python
nrv_pwl_df = None
for depdate in depdates:
    fname = "s3://ay-emr-job/nrm/nrv_pwl/{}/{}/{}/nrv_pwl_{}_{}.csv.gz".format(fcstdate[:4], fcstdate[4:6], fcstdate[6:8], fcstdate, depdate)
    df = pd.read_csv(fname, low_memory = False)
    if df is None:
        nrv_pwl_df = df 
    else:
        nrv_pwl_df = pd.concat([nrv_pwl_df, df])
```
