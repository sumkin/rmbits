```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../data/data.csv")
```

```python
df["profit_fixed"] = df["revenue_fixed"] - df["costs_fixed"]
df["profit_partial"] = df["revenue_partial"] - df["costs_partial"]
df["profit_optimized"] = df["revenue_optimized"] - df["costs_optimized"]

df["revenue_percentage"] = (df["revenue_partial"] - df["revenue_fixed"]) / df["revenue_fixed"]
df["costs_percentage"] = (df["costs_partial"] - df["costs_fixed"]) / df["costs_fixed"]
df["profit_percentage"] = (df["profit_partial"] - df["profit_fixed"]) / abs(df["profit_fixed"])
```

```python
df[["fcstdate","profit_percentage"]].head(20)
```

```python
df["fcstdate"] = df["fcstdate"].astype(str)
```

```python
fig = px.line(df, x = "fcstdate", y = ["profit_fixed", "profit_partial", "profit_optimized"])
fig.show()
```
